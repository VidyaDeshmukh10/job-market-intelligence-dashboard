```python
import pandas as pd
import sqlite3
```

```python
jobs = pd.read_csv("C:/Users/lenovo/OneDrive/Desktop/Projects/Job Market Intelligence/data/processed/jobs_clean.csv")
```

```python
conn =sqlite3.connect("../sql/job_market.db")
```

```python
jobs.to_sql(
    "jobs",
    conn,
    if_exists = "replace",
    index = False
)
```

1. View the dataset

```python
pd.read_sql(
    "SELECT * FROM jobs LIMIT 5;",
    conn
)
```

2. Top 20 Hiring Companies

```python
pd.read_sql(
    """
    SELECT company_name, COUNT(*) AS total_jobs
    FROM jobs
    GROUP BY company_name
    ORDER BY total_jobs DESC
    LIMIT 20;
""",
conn
)
```

3. Average Salary by Experience Level

```python
pd.read_sql(
    """
    SELECT
    experience_level,
    AVG(salary_standardized) AS average_salary
FROM jobs
GROUP BY experience_level
ORDER BY average_salary DESC;
""",
    conn
)
```

4. Jobs with Salary Greater than $100,000

```python
pd.read_sql(
    """
    SELECT *
    FROM jobs
    WHERE salary_standardized > 100000;
""",
    conn
)
```

5. Senior Jobs with Salary Greater than $100,000

```python
pd.read_sql(
    """
    SELECT *
FROM jobs
WHERE experience_level = 'Senior'
AND salary_standardized > 100000;
""",
    conn
)
```

6. Intern or Junior Jobs

```python
pd.read_sql(
    """
    SELECT *
FROM jobs
WHERE experience_level = 'Intern'
OR experience_level = 'Junior';
""",
    conn
)
```

7. Top 5 Companies Offering Senior Jobs with Salary > $100,000

```python
pd.read_sql(
    """
    SELECT
    company_name,
    COUNT(*) AS total_jobs
FROM jobs
WHERE experience_level = 'Senior'
AND salary_standardized > 100000
GROUP BY company_name
ORDER BY total_jobs DESC
LIMIT 5;
""",
    conn
)
```

8: Top Hiring Locations

```python
pd.read_sql(
    """
    SELECT 
    location,
    COUNT(*) AS total_jobs
    FROM jobs 
    GROUP BY location
    ORDER BY total_jobs DESC
    LIMIT 20;
    """,
    conn
)
```

9: Experience Level Distribution

```python
pd.read_sql(
    """
    SELECT 
    experience_level,
    COUNT(*) AS total_jobs
    FROM jobs
    GROUP BY experience_level
    ORDER BY total_jobs DESC
    """,
    conn
)
```

10: Salary Statistics

```python
pd.read_sql(
    """
    SELECT 
    MIN(salary_standardized) AS minimum_salary,
    MAX(salary_standardized) AS maximum_salary,
    AVG(salary_standardized) AS average_salary
    FROM jobs;
    """,
    conn

)
```

11: Remote Jobs Only

```python
pd.read_sql(
    """
    SELECT *
    FROM jobs
    WHERE work_from_home = 1;
    """,
    conn

)
```

Companies Starting with "A" (LIKE)

```python
pd.read_sql(
    """
    SELECT *
    FROM jobs
    WHERE company_name LIKE "A%"
    """,
    conn

)
```

13: Salary Between 80K and 120K (BETWEEN)

```python
pd.read_sql(
    """
    SELECT *
    FROM jobs
    WHERE salary_standardized BETWEEN 80000 AND 120000;
    """,
    conn
)
```

14: Experience Level Using IN

```python
pd.read_sql(
    """
    SELECT * 
    FROM jobs
    WHERE experience_level IN ("Intern","Junior","Mid");
    """,
    conn
)
```

15: Companies with More Than 100 Job Postings (HAVING)

```python
pd.read_sql(
    """
    SELECT  
    company_name,
    COUNT(*) AS total_jobs
    FROM jobs
    GROUP BY company_name
    HAVING COUNT(*)>100
    ORDER  BY total_jobs DESC;
    """,
    conn
)
```
